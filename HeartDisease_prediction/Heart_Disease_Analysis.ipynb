### Combination Of Different Columns Data

```python
# Convert columns to appropriate data types
ht['trestbps'] = pd.to_numeric(ht['trestbps'], errors='coerce')
ht['chol'] = pd.to_numeric(ht['chol'], errors='coerce')
```

```python
ht[(ht['cp'] == 'typical angina') & (ht['trestbps'] > 130) & (ht['chol'] > 200)]
```

```python
# Visualization of Blood pressure and cholestrol
ht[['trestbps', 'chol']].plot(kind='bar')
```

```python
# Resting ECG and Maximum Heart Rate
ht.groupby('restecg')['thalch'].mean()
```

```python
# Viausalization
ht.groupby('restecg')['thalch'].mean().plot(kind='bar')
```

```python
# Exercise-Induced Angina
ht.groupby('exang')['num'].value_counts(normalize=True)
```

```python
# Visualization
ht.groupby('exang')['num'].value_counts(normalize=True).unstack().plot(kind='bar', stacked=True)
```

```python
# ST Depression and Slope
ht.groupby('slope')['oldpeak'].mean()
```

```python
# Visualization
ht.groupby('slope')['oldpeak'].mean().plot(kind='line')
```

```python
# Number of Major Vessels and Thalassemia
ht.groupby(['ca', 'thal'])['num'].value_counts(normalize=True)
```

```python
# Visualization
ht.groupby(['ca', 'thal'])['num'].value_counts(normalize=True).unstack().plot(kind='bar', stacked=True)
```
